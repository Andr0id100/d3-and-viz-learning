## Viz Work

```python
with open("slam-data/es_en.slam.20190204.dev", "r") as f:
    raw_data = f.read()
```

```python
with open("slam-data/es_en.slam.20190204.dev.key", "r") as f:
    raw_success_data = f.read()
```

```python
success_key = {}
for x in raw_success_data.split("\n"):
    entry = x.split()
    if len(entry) == 2:
        success_key[entry[0]] = eval(entry[1])
```

```python
lines = raw_data.split("# ")[2:] #Removing the empty one and the promptless one
```

```python
len(lines), lines[:10]
```

```python
# Structure of processed data
# (Prompt, User-ID, Days, Success-Status)
```

```python
data = set()
prompt = None
for x in lines:
    x = x.strip()
    if "prompt:" in x:
        prompt = x[len("prompt:"):]
    else:
        entries = x.split()
        user_id = entries[0][len("user:+"):]
        days = eval(entries[2][len("days:"):])
        status = success_key[entries[7]]
#         print(prompt, user_id, days, status)
        data.add((prompt, user_id, days, status))
```
